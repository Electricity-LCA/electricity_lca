```python
from entsoe import EntsoePandasClient
import pandas as pd
import polars as pl
import dotenv
import datetime
import os
```

```python
dotenv.load_dotenv('.env')
ENTSOE_SECURITY_TOKEN = os.getenv('ENTSOE_SECURITY_TOKEN')
client = EntsoePandasClient(ENTSOE_SECURITY_TOKEN)
```

## Country 1 : The Netherlands (NL)

```python
start_date = pd.Timestamp('2023-12-21 00:00:00+0:00')
end_date = pd.Timestamp('2023-12-31 00:00:00+0:00')
```

```python
generation_nl = client.query_generation('NL',start=start_date,end=end_date)
```

```python
generation_nl
```

```python
import plotly.graph_objects as plgo
import plotly.express as plx
generation_type = 'Fossil Gas'
country_code = 'NL'
fig = plx.line(generation_nl.loc[:,pd.IndexSlice[generation_type]])
fig.layout = plgo.Layout({'title':f'{generation_type} in {country_code} (MW)'})
fig.show()
```
